```python
import pandas as pd 
```

```python
df = pd.read_csv("dataset.csv")
display(df.info())
```

```python
# Convert coordinate columns to float
df['ocorrencia_latitude'] = pd.to_numeric(df['ocorrencia_latitude'], errors='coerce')
df['ocorrencia_longitude'] = pd.to_numeric(df['ocorrencia_longitude'], errors='coerce')
df.info()
```

```python
# remover a coluna "divulgacao_dia_publicacao" com "232 non-null" tipo "object" 
df.drop(columns=["divulgacao_dia_publicacao"], inplace=True)
# dropar os valores que sao nulos 
df.dropna(inplace=True)
# dropar as colunas identificadoras 
df.drop(
    columns=[
        "codigo_ocorrencia",
        "codigo_ocorrencia1", 
        "codigo_ocorrencia2", 
        "codigo_ocorrencia3", 
        "codigo_ocorrencia4", 
        "aeronave_matricula",
    ], 
    inplace=True
)
# dropar colunas com valores unicos
df.drop(
    columns=[
        "aeronave_operador_categoria", 
        "ocorrencia_pais"
    ], 
    inplace=True
)
df.info()
```

```python
# Remove rows that contain "***" in any cell
display(df[~df.isin(['***']).any(axis=1)])

## PORTANTO NAO eh possivel remover as linhas com "***"
```

```python
from pareto_discretizer import pareto_discretize

df = pareto_discretize(df, threshold=0.8, drop_others_dominant=True, exclude_columns=["ocorrencia_classificacao"])

print(df.shape)

for col in df.columns:
    print(f"Coluna: {col}")
    display(df[col].value_counts())
    print("\n")
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Get categorical columns, excluding "ocorrencia_classificacao"
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()
if "ocorrencia_classificacao" in categorical_columns:
    categorical_columns.remove("ocorrencia_classificacao")

# Apply one-hot encoding
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_features = encoder.fit_transform(df[categorical_columns])

# Get feature names
feature_names = encoder.get_feature_names_out(categorical_columns)

# Create DataFrame with encoded features
encoded_df = pd.DataFrame(encoded_features, columns=feature_names, index=df.index)

# Get numerical columns and add "ocorrencia_classificacao" if it exists
numerical_columns = df.select_dtypes(exclude=['object', 'category']).columns.tolist()
non_encoded_columns = numerical_columns.copy()
if "ocorrencia_classificacao" in df.columns:
    non_encoded_columns.append("ocorrencia_classificacao")

# Combine with numerical columns and "ocorrencia_classificacao"
df_encoded = pd.concat([df[non_encoded_columns], encoded_df], axis=1)

print(f"Original shape: {df.shape}")
print(f"After one-hot encoding shape: {df_encoded.shape}")
print(f"Categorical columns encoded: {categorical_columns}")

df = df_encoded.reset_index(drop=True)
display(df)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Get numerical columns (excluding categorical encoded columns and target)
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Remove target column if it exists
if "ocorrencia_classificacao" in numerical_columns:
    numerical_columns.remove("ocorrencia_classificacao")

# Apply MinMax scaling to numerical columns
scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[numerical_columns] = scaler.fit_transform(df[numerical_columns])

print(f"Numerical columns scaled: {numerical_columns}")
print(f"Shape after scaling: {df_scaled.shape}")

df = df_scaled
display(df)
```

```python
df = pd.read_csv("dataset-preprocessed.csv")
df["ocorrencia_classificacao"].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Separate features and target
X = df.drop('ocorrencia_classificacao', axis=1)
y = df['ocorrencia_classificacao']

# Apply SMOTE to balance the classes
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create balanced dataframe
df_balanced = pd.DataFrame(X_resampled, columns=X.columns)
df_balanced['ocorrencia_classificacao'] = y_resampled

print(f"Original class distribution:")
print(y.value_counts())
print(f"\nBalanced class distribution:")
print(y_resampled.value_counts())
print(f"\nOriginal shape: {df.shape}")
print(f"Balanced shape: {df_balanced.shape}")

# Update df to the balanced version
df = df_balanced
display(df)
df.to_csv("dataset-preprocessed-balanced-smote.csv", index=False)
```
